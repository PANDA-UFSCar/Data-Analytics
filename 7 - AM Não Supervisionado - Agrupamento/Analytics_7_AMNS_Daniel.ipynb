#Diferenças entre os tipos de métodos

##Particionais vs Hierarquicos

###Particional


Definição:
*   Constroem uma partição dos dados


Número de Clusters:
*   Nos métodos particionais, você precisa especificar o número de clusters que deseja formar antes de iniciar o processo de agrupamento.
Processo de Atribuição:

Processo de Atribuição:
*   O processo de agrupamento envolve atribuir cada ponto de dados a um cluster com base em uma métrica de distância ou similaridade. Um ponto pode pertencer apenas a um cluster.

Algoritmos Comuns:
*   K-Means, K-Medoids, Fuzzy C-Means

Escalabilidade:
*   Geralmente, os métodos particionais são mais escaláveis para conjuntos de dados grandes, uma vez que a atribuição de pontos a clusters pode ser calculada de maneira mais eficiente.

Resultado Sensível à Inicialização:
*   Os resultados do agrupamento podem depender da inicialização dos centroides ou pontos de referência iniciais.


###Hierarquico

Definição:
*   Constroem uma hierarquia de partições

Hierarquia de Clusters:

*   Os métodos hierárquicos não requerem o número de clusters a priori. Eles criam uma hierarquia de clusters, começando com cada ponto como um cluster individual e, em seguida, combinando gradualmente clusters em níveis superiores.

Processo de Fusão ou Divisão:
*   Os métodos hierárquicos podem ser **aglomerativos** (começam com clusters individuais e os combinam) ou **divisivos** (começam com um único cluster contendo todos os pontos e os dividem em clusters menores).

Algoritmos Comuns:
*   Agglomerative Clustering, Divisive Clustering e Complete Linkage Clustering.

Visualização de Dendrogramas:
*   Uma característica importante dos métodos hierárquicos é a possibilidade de representar o resultado por meio de um dendrograma, que mostra a estrutura de hierarquia de clusters.

Menos Sensível à Inicialização:
*   Os métodos hierárquicos geralmente são menos sensíveis à inicialização em comparação com os métodos particionais.

Escalabilidade Limitada:
*   À medida que o tamanho do conjunto de dados aumenta, a escalabilidade dos métodos hierárquicos pode se tornar um desafio devido ao cálculo de distâncias e à construção da estrutura hierárquica.

##Relacionais vs Não-Relacionais

###Relacional

Definição:
*   Requerem apenas as (dis)similaridades entre os objetos

Leva a:
*   Abordagem unificada para tramaneto de atributos mistos
*   Dados Sigilosos
*   Mas também, a um custo computacional muito elevado

Dados Estruturados:
*   Lidam com dados organizados em tabelas com relações definidas.

Métodos Clássicos:
*   Utilizam técnicas estatísticas tradicionais, como regressão linear, ANOVA, t-testes, etc.

Algoritmos Comuns:
*   K-means, PCA, Hierarchical Clustering

Interpretação Direta:
*  Resultados podem ser facilmente interpretados devido à estrutura organizada dos dados.


###Não-Relacional

Definição:
*   Requerem os objetos a serem agrupados (dados originais)

Dados Não Estruturados ou Semi-Estruturados:
*   Lidam com dados que não se encaixam bem em uma estrutura tabular, como textos, imagens, séries temporais.

Métodos Avançados:
*   Requerem técnicas estatísticas mais avançadas, como aprendizado de máquina e análise de séries temporais, para lidar com a complexidade dos dados.

Algoritmos Comuns:
*   DBSCAN, MeanShifting

Desafios de Interpretação:
*   Resultados podem ser mais desafiadores de interpretar devido à natureza não estruturada dos dados.

##Similaridade vs Densidade

###Similaridade

Base de Funcionamento:
*   Os métodos de similaridade agrupam os pontos de dados com base em sua proximidade ou similaridade entre si.

Critério de Agrupamento:
*   Os algoritmos de similaridade geralmente separam pontos mais semelhantes em grupos, onde a similaridade é medida usando métricas de distância ou outras medidas de proximidade.

Exemplo de Algoritmo:
*   O K-Means é um exemplo de algoritmo de similaridade que tenta dividir os pontos de dados em clusters de modo que a distância intra-cluster seja minimizada.

Adequado para:
*   Dados onde os clusters têm formas bem definidas e podem ser separados por fronteiras claras.

Desafios:
*   Pode ser menos eficaz quando os clusters têm formas complexas ou tamanhos variáveis, pois a distância pode não ser uma medida adequada de similaridade.


###Densidade

Base de Funcionamento:


*   Os métodos de densidade procuram identificar áreas densas de pontos no espaço de dados.




Critério de Agrupamento:
*   Eles identificam regiões com alta densidade de pontos e consideram essas regiões como clusters.

Exemplo de Algoritmo:
*   O DBSCAN é um exemplo de algoritmo de densidade que agrupa pontos que estão próximos uns dos outros e têm densidade suficientemente alta.

Adequado para:
*   Dados onde os clusters têm formas irregulares e podem não ser separados por fronteiras bem definidas.

Desafios:
*   Pode ser sensível a parâmetros, como o tamanho da vizinhança, e pode ter dificuldade em lidar com clusters de densidades diferentes.


#Algoritmos comuns

##Single Linkage

O método Single Linkage, também conhecido como método de ligação única ou método de ligação mínima, é uma abordagem específica dentro do clustering aglomerativo.

Ele determina a similaridade entre dois clusters (grupos de pontos de dados) com base na distância mínima entre um ponto de um cluster e um ponto de outro cluster.

Em outras palavras, ele calcula a distância entre os pontos mais próximos dos clusters para definir a similaridade entre eles.

Aqui está um passo a passo de como o método Single Linkage funciona:




1.   É feito uma matriz de distância entre todos os elementos (pares)
2.   Localiza os dois elementos ou cluster com a menor distância entre si [Pode se utilizar Distancias Euclidianas ou Manhattan]
3.   Faz a fusão os dois elementos ou clusters em um só
4.   Atualiza a Matriz de distâncias
5.   Repete os passos 2-3-4 até que todos estejam em um só grupo(cluster)





##K-MEANS


Passo a passo:


1.   Escolha o Número de Clusters (k)
Determine quantos clusters você deseja identificar nos seus dados. Isso pode ser baseado em conhecimento prévio ou usando métodos como o método do cotovelo.

2.   Definir uma centróide para cada cluster de forma aleatoria

3.   Calcular, para cada ponto, o centroide de menor distância.

4.   Reposicionar o centro de cada cluster, com a média da posição de todos os pontos do cluster.

5.   Repetir os passos 3 e 4 até obter um centroide satisfatorio.



Vantagens:
*   É rápido e eficiente em grandes conjuntos de dados.
*   É fácil de entender e implementar.

Desvantagens:
*   Requer que o número de clusters seja conhecido a priori.
*   Não funciona bem com clusters de formas complexas ou tamanhos diferentes.



##DBSCAN

Algoritmo baseado na densidade, para tal, possui 2 parâmetros bases:


*   Epsilon:
Raio máximo em que dois pontos podem estar para serem considerados do mesmo cluster.

*   Quantidade minima de pontos:
Número mínimo de pontos em uma região necessários para garantir uma densidade desejada.

Outros conceitos importantes:

*   Núcleos: Um ponto é considerado núcleo quando possui a quantidade mínima de pontos ou mais dentro de um raio de Eps de distância.
*   Bordas: Um ponto é considerado borda quando não possui a quantidade de pontos mínima dentro de um raio Eps de distância, porém está inserido dentro de um raio Eps de um outro ponto no qual é núcleo.
*   Ruídos(ou Outliers): Um ponto é considerado um ruído quando este não está presente no raio Eps de um núcleo e nem possui MinPts dentro do seu raio Eps.

Dessa forma, o passo a passo seria:



1.   Escolher um ponto arbitrário
2.   Classificá-lo como Núcleo, Borda ou Ruído
3.   Se o ponto do momento é um Núcleo, todos os pontos à sua volta com um raio de distância Eps formam um cluster. Reclassificações podem ocorrer.
4.   Depois de classificar todos os pontos da região do núcleo, volta para a etapa 1.




Vantangens:
*   Utilizar DBSCAN não requer a pré definição de números de clusters e utiliza apenas dois parâmetros.
*   Os parâmetros são facilmente ajustados caso haja um bom conhecimento sobre os dados.
*   DBSCAN se comporta muito bem em identificar clusters de tamanhos arbitrários, como visto no exemplo de implementação.
*   A classificação de ruídos é muito bem feita.

Desvantagens:
*   DBSCAN não performa bem em datasets com variabilidade de densidade entre os dados.
*   O DBSCAN não é uma técnica completamente determinística. Isto é possível de perceber tratando de pontos de borda que estão dentro do raio de núcleos referentes a diferentes clusters. Na maioria das situações, pode ser que isso não influencie em uma análise.

##Davies Bouldin Score

 Índice de Davies-Bouldin é uma métrica de validação de clusters que é usada para avaliar a qualidade dos clusters obtidos após a aplicação de um algoritmo de clustering, como o K-Means. O objetivo do DBI é medir a separação e a compacidade dos clusters, isto é, compacidade é um conceito que descreve o grau em que os elementos ou objetos em um grupo ou conjunto estão próximos uns dos outros em relação a uma métrica de distância específica. Dessa forma, um valor menor do índice indica uma melhor separação e compacidade dos clusters. Quanto mais próximo de zero, melhor.

##Jacard Score

Use a fórmula do Jaccard Score para calcular a similaridade entre os conjuntos A e B:

J(A,B)=∣A∩B∣ / ∣A∪B∣

Isso lhe dará um valor entre 0 e 1, onde 0 indica nenhuma similaridade (os conjuntos são completamente diferentes) e 1 indica completa similaridade (os conjuntos são idênticos). Quanto maior o valor do Jaccard Score, maior a similaridade entre os conjuntos.

##Dunn Score

O Dunn Index, também conhecido como Índice de Dunn, é uma métrica de validação de clusters que é usada para avaliar a qualidade dos clusters obtidos após a aplicação de um algoritmo de clustering, como o K-Means. O objetivo do Dunn Index é medir a separação entre os clusters e a compacidade dentro dos clusters, sendo que um valor maior do índice indica uma melhor separação e uma melhor compacidade dos clusters. Em outras palavras, ele mede a relação entre o tamanho dos clusters e a distância entre eles.
