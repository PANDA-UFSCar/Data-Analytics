###Comparação entre regiões

````python
from plotly.subplots import make_subplots

south_data = df[df["Region"] == "South"]
west_data = df[df["Region"] == "West"]
central_data = df[df["Region"] == "Central"]
east_data = df[df["Region"] == "East"]

total_profit_south = south_data["Profit"].mean()
total_profit_west = west_data["Profit"].mean()
total_profit_central = central_data["Profit"].mean()
total_profit_east = east_data["Profit"].mean()


total_sales_south = south_data["Sales"].mean()
total_sales_west = west_data["Sales"].mean()
total_sales_central = central_data["Sales"].mean()
total_sales_east = east_data["Sales"].mean()


total_dsc_south = south_data["Discount"].mean()
total_dsc_west = west_data["Discount"].mean()
total_dsc_central = central_data["Discount"].mean()
total_dsc_east = east_data["Discount"].mean()

colors = ['rgb(158,202,225)']

fig = make_subplots(rows=1, cols=3, subplot_titles=("Lucro", "Vendas","Desconto"))

profit_bar_trace = go.Bar(x=['south', 'west', 'central', 'east'], y=[total_profit_south, total_profit_west, total_profit_central, total_profit_east],
            marker_color=['#7B68EE','#8A2BE2','#9370DB','#7B68EE'], marker_line_color=['rgb(8,48,107)','rgb(20,0,07)'],
                  marker_line_width=1.5, opacity=1,
            hovertext=['', 'Região com mais lucro', 'Região com menos lucro'], name="Lucros por região")

sales_bar_trace = go.Bar(x=['south', 'west', 'central', 'east'], y=[total_sales_south, total_sales_west, total_sales_central, total_sales_east],
            marker_color=['#8A2BE2','#9370DB','#7B68EE','#7B68EE'], marker_line_color=['rgb(8,48,107)','rgb(20,0,07)'],
                  marker_line_width=1.5, opacity=1,
            hovertext=['Região com menos vendas', 'Região com mais vendas', ''], name="Vendas por região")

discount_bar_trace = go.Bar(x=['south', 'west', 'central', 'east'], y=[total_dsc_south, total_dsc_west, total_dsc_central, total_dsc_east],
            marker_color=['#7B68EE','#9370DB','#8A2BE2','#7B68EE'], marker_line_color=['rgb(8,48,107)','rgb(20,0,07)'],
                  marker_line_width=1.5, opacity=1,
            hovertext=['Região com menos descontos', '', 'Região com mais descontos'], name="Desconto por região")


fig.add_trace(profit_bar_trace, row=1, col=1)
fig.add_trace(sales_bar_trace, row=1, col=2)
fig.add_trace(discount_bar_trace, row=1, col=3)




# Customize aspect

fig.update_layout(title_text='Em cada região temos',height=500, width=1500,
                  font_family="Courier New",
                  font_color="white",
                  font_size=15,
                  #Mudar o background
                  paper_bgcolor="black",
    #Muda o bckground do grafico
                  plot_bgcolor='black',)

fig.show()
````

O que podemos tirar disso:


*   A reigão West tem o maior lucro e se encontra também em uma das regiões com os menores descontos
*   A região Sul tem uma das maiores vendas
*   A região central tem pouco lucro e também possui um alto desconto, será por isso uma da possibilidade de seu problema

#Conclusão

Areas fracas onde o empresário pode trabalhar para fazer mais lucro:


*   Vendas e Lucro não são linear na maioria dos estados americanos
*   A região central necessita de mais atenção
*   As categorias de mobília e escritório tem os piores lucros, por conta do alto desconto
*   A categoria de escritório tem uma perda máxima de até 80% com 0% de lucro
*   A categoria de mobília e tecnologia, tem uma perda máxima entre 40% e 50%
*   Investimentos em mesas e estantes de livros, tem lucro negativo


Problemas de negócios que você pode encontrar explorando esses dados:


*   Tecnologia lucra mais se comparado com mobílias
*   No tipo de frete 'Same Day' tem as maiores vendas
*   A região Oeste tem o maior lucro
*   O estado de Vermont tem um bom lucro com poucas vendas
*   É necessário dar mais desconto na categoria de mobília para atrair mais consumidores
