Desvantagens:

O pacote pyplot serve para a construção e customização de gráficos.

```python
import matplotlib.pyplot as plt
```

# Gráfico de linhas

```python
x = [10, 15, 20, 25]
y = [1, 5, 10, 50]
plt.plot(x,y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico')

plt.show()
```

Modificando a legenda e a cor do gráfico:

```python
x1 = [10, 15, 20, 25]
y1 = [1, 5, 10, 50]

x2 = [15, 40, 80, 100]
y2 = [5, 12, 30, 50]

plt.plot(x1, y1, label = 'Gráfico 1', color = 'Green')
plt.plot(x2, y2, label = 'Gráfico 2', color = 'Purple')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico')
plt.legend()

plt.show()
```

# Gráfico de barras

```python
x1 = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]

plt.bar(x1, y1)
plt.title('Gráfico de barras')
plt.show()
```

```python
x2 = ['A', 'B', 'C', 'D']
y2 = [5, 10, 15, 20]

plt.bar(x2, y2, color='Purple')
plt.title('Gráfico de barras')
plt.show()
```

```python
plt.bar(x2, y1, color = 'Red', label = 'Lista 1')
plt.bar(x2, y2, color = 'Blue', label = 'Lista 2')
plt.title('Gráfico de barras empilhadas')
plt.legend()
plt.show()
```

```python
x3 = [1, 5, 8, 12]
y3 = [10, 15, 5, 20]

x4 = [3, 6, 9, 11]
y4 = [5, 10, 15, 20]

plt.bar(x3, y3, label = 'Lista 1')
plt.bar(x4, y4, label = 'Lista 2')
plt.legend()
plt.show()
```

# Gráfico de dispersão

```python
x = [5, 14, 17, 12, 10, 16, 20, 25]
y = [4, 8, 5, 16, 7, 10, 6, 12]

plt.scatter(x, y)
plt.title('Gráfico de dispersão')
plt.show()
```

# Gráfico de áreas empilhadas

```python
a = [15, 17, 19, 20]
b = [6, 8, 10, 12]
c = [10, 20, 30, 40]
d = [5, 10, 15, 20]

plt.stackplot(a, b, c, d, labels = ['a', 'b', 'c', 'd'])
plt.legend()
plt.show()
```

# Gráfico de pizza

```python
labels = ['Panda', 'Cati Jr', 'Red Dragons', 'Pet']
colors = ['Purple', 'Royalblue', 'Red', 'Green']
quantidade = [10, 3, 4, 6]
plt.pie(quantidade, labels = labels, colors = colors, autopct='%1.1f%%') #startangle, explode
plt.show()
```

# Histograma

```python
import numpy as np
x = np.random.randn(100)

fig, axes = plt.subplots(1, 2) #1 linha 2 colunas

axes[0].hist(x)
axes[0].set_xlim(min(x), max(x)) #ajusta os limites do eixo x
axes[0].set_title('Histograma padrão')

axes[1].hist(x, cumulative = True, bins = 50)
axes[1].set_title('Histograma cumulativo')
```

## Seaborn


Vantagens:
*   Possibilidade de criar gráficos com comandos simples
*   Pode usar estruturas de dados do pandas como input

---
Desvantagens:
*   Limitações na customização
*   Visualizações não tão interativas




```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
dados = sns.load_dataset('taxis')
dados
```

# Jointplot

```python
sns.jointplot(data = dados, x = 'distance', y = 'total', kind = 'reg')
```

#Lmplot

```python
sns.lmplot(data = dados, x = 'distance', y = 'total', col = 'passengers')
```

# Matriz de dispersão

```python
sns.pairplot(dados)
```

# Kdeplot

```python
sns.kdeplot(dados['passengers'])
```

#Boxplot

```python
sns.boxplot(data = dados, x = 'passengers', y = 'tip')
plt.xlabel('Número de passageiros')     #juntar sns com plt
plt.ylabel('Quantidade de gorgeta')
plt.show()
```

#Violinplot

```python
sns.violinplot(data = dados,x = 'passengers', y='distance')
```

#Gráfico de barras GO

```python
import plotly.graph_objects as go
x = [10, 20, 30, 40]
y = [5, 10, 15, 20]

fig = go.Figure()
fig.add_trace(go.Bar(x = x, y = y))
fig.update_layout(
    template = 'plotly_dark',
    title = dict(
        text = 'Gráfico de Barras',
        x = 0.5,
        font  = dict(
            family = 'Arial',
            size = 20
        )
    ),
    xaxis_title = 'Eixo X',
    yaxis_title = 'Eixo Y'

)
fig.show()
```

#Boxplot GO

```python
fig = go.Figure()
fig.add_trace(go.Box(x = dados['passengers'], y = dados['distance'], fillcolor = 'purple', marker_color = 'green'))
fig.update_layout(
    template = 'plotly_dark',
    title = dict(
        text = 'Boxplot',
        x = 0.5,
        font = dict(
            family = 'Arial',
            size = 30
        )
    ),
    xaxis_title = 'Número de passageiros',
    yaxis_title = 'Distância percorrida',
    #boxfillcolor='lightblue'
)

```
