## Propriedade das cores

- Matiz — também chamado de tom (ou tonalidade), corresponde a própria cor, e.g., vermelho, amarelo e azul.
- Luminosidade — também chamado de valor (ou valor tonal), corresponde a “claridade” da cor, i.e., se uma cor é mais clara ou mais escura (quantidade de luz e sombra).
- Saturação — também chamado de intensidade, corresponde a “pureza” da cor, i.e., se uma cor é mais “forte” (”mais pura”) ou mais próxima de cinza (”menos pura”).

## Temperatura das cores

- Cores quentes.
- Cores frias.

## Classificação das cores

- Cores Primárias — também chamadas de cores puras, fazem parte desse grupo o vermelho, azul e amarelo.
    - Definição: cores que não podem ser decompostas em outras cores, mas criam outras quando combinadas.
- Cores Secundárias — compostas pela junção de duas cores primárias, e.g., as cores laranja, verde e roxo.
- Cores Terciárias — mistura de uma cor primária com uma secundária, como o vermelho arroxeado.
    - Duas cores não cabem nessa classificação: o preto e branco. Já que o branco consiste na junção de todas as cores, enquanto o preto corresponde a ausência de cor.

## Síntese de cores

- Síntese aditiva — observa-se a luz emitida por um objeto, tendo como seu sistema representativo o RGB (*Reg Green Blue*).
- Síntese subtrativa — observa-se a luz refletida por um objeto, tendo como sistemas representativos CYMK (*Cyan Yellow Magenta Key*) e RYB (*Red Yellow Blue*).

## Harmonia das cores

- Harmonia monocromática — uma cor inicial é escolhida e trabalha-se sobre ela (variações de de tonalidade.
    - Simples.
    - Contraste entre cores não é importante.
    - Equilibro entre tonalidades.
- Harmonia análoga — escolhe-se uma cor primário que é usado junto com aquelas próximas a ela na roda de cores.
    - Cor primária → principal.
    - Cores seguintes → aumentam a harmonia do conjunto (detalhes e complementos).
    - Mais contraste e diversidade de combinações comparado ao monocromático.
- Harmonia complementar — escolhe-se uma cor predominante que é usada junto com a cor complementar correspondente.
    - Utiliza-se duas cores complementares (i.e., com mais contraste), que estão em pontos opostos da roda de cores.
    - Ressalta o contraste de cores de forma mais evidente.
- Harmonia triádica — escolhe-se três cores equidistantes na roda de cores.
    - Equilíbrio entre contraste de cores e harmonia do conjunto.
    - Nota: é interessante tem uma das cores como a principal e as outras para efeitos secundários.

## Psicologia das cores

Área de estudo que pesquisa como as cores são capazes de motivar sentimentos, desejos e emoções. Sua função é fazer um mapeamento das simbologias mais prováveis que podem ser assimiladas quando uma determinada cor é utilizada.

### Gráfico de linha

```python
import plotly.express as px

df = px.data.gapminder().query("continent=='Europe'")
fig = px.line(df, x="year", y="lifeExp", color='country')
fig.show()
```

### Gráfico de dispersão

Dados são representados como pontos com coordenadas correspondentes aos valores de dois atributos numéricos.

Útil para visuzalizar possível correlação entre um par de atributos.

```python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()
```

### Gráfico de bolhas

Derivado do gráfico de dispersão, essa representação considera-se um atributo adicional, o que aumenta a dimensionalidade do gráfico.

Pontos passam a ser círculos com área proporcional ao valor do atributo adicional.

```python
df = px.data.gapminder()

fig = px.scatter(df.query("year==2007"), x="gdpPercap", y="lifeExp",
	               size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)
fig.show()
```

### Gráfico de pontos

Derivado do gráfico de dispersão, utiliza um atributo numérico e outro nominal.

```python
schools = ["Brown", "NYU", "Notre Dame", "Cornell", "Tufts", "Yale",
           "Dartmouth", "Chicago", "Columbia", "Duke", "Georgetown",
           "Princeton", "U.Penn", "Stanford", "MIT", "Harvard"]
n_schools = len(schools)

women_salary = [72, 67, 73, 80, 76, 79, 84, 78, 86, 93, 94, 90, 92, 96, 94, 112]
men_salary = [92, 94, 100, 107, 112, 114, 114, 118, 119, 124, 131, 137, 141, 151, 152, 165]

df = pd.DataFrame(dict(school=schools*2, salary=men_salary + women_salary,
                       gender=["Men"]*n_schools + ["Women"]*n_schools))

# Use column names of df for the different parameters x, y, color, ...
fig = px.scatter(df, x="salary", y="school", color="gender",
                 title="Gender Earnings Disparity",
                 labels={"salary":"Annual Salary (in thousands)"} # customize axis label
                )

fig.show()
```

### Gráfico de pizza

Círculo divido em setores que ilustram suas respectivas proporções numéricas.

Recomenda-se ter uma quantidade pequena de setores, a fim de facilitar sua leitura e evitar que fique poluído.

```python
import plotly.express as px

df = px.data.tips()
fig = px.pie(df, values='tip', names='day')
fig.show()
```

### Gráfico de explosão solar

Semelhante ao gráfico de pizza, entretanto, plota dados de forma hierarquica partindo da raiz (círculo central) para as folhas de forma radial.

A raíz começa no centro e seus filhos são adcionados aos aneis externos.

A biblioteca permite que um filho seja para que seja tratado como uma raíz.

```python
import plotly.express as px

data = dict(
    character=["Eve", "Cain", "Seth", "Enos", "Noam", "Abel", "Awan", "Enoch", "Azura"],
    parent=["", "Eve", "Eve", "Seth", "Seth", "Eve", "Eve", "Awan", "Eve" ],
    value=[10, 14, 12, 10, 2, 6, 6, 4, 4])

fig = px.sunburst(
    data,
    names='character',
    parents='parent',
    values='value',
)
fig.show()
```

### Gráfico de barras

Cada atributo de uma coluna é representado como uma barra, que tem seu tamanho correspondente a soma de outro atributo dado aquele da barra.

```python
import plotly.express as px

df = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(df, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], color='lifeExp',
             labels={'pop':'population of Canada'}, height=400)
fig.show()
```

### Box plot

Representação estatística da distribuição de uma variável por meio de seus quartis. A caixa é delimitada pelos quartis inferior e superior, enquanto o traço em seu interior é a mediana.

Os "bigodes" são delimitados pelos valores máximo ou mínimo caso esteja dentro do intervalo baseado na amplitude interquartil, caso contrário, correspondem aos limites desse intervalo e os valores fora dele são marcados como pontos.

```python
df = px.data.tips()
fig = px.box(df, x="time", y="total_bill")
fig.show()
```

### Histograma

Representação da distribuição numérica dos dados, em que os dados são separados em <i>bins</i> que são representados como a contagem de seus valores.

Utilizado para visualizar a distribuição de atributos contínuos.

```python
df = px.data.tips()
fig = px.histogram(df, x="total_bill")
fig.show()
```

### Gráfico de distribuição

Combinação de um histograma com a estimativa de densidade kernel dos dados.

```python
import plotly.figure_factory as ff

np.random.seed(1)

x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot'] # name of the dataset

fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```

### Histograma 2D

Também chamado de mapa de calor de densidade, correponde a generalização em 2 dimensões do histograma. Embora se assemelhe ao mapa de calor, essa representação faz seu cálculo agrupando pontos específicos por meio das "coordenadas" dos <i>bins</i> e da aplicação de alguma função, como contagem ou soma, então uma cor é associada para representar o <i>bin</i> com base nesse valor.

```python
df = px.data.tips()

fig = px.density_heatmap(df, x="total_bill", y="tip", nbinsx=20, nbinsy=20)
fig.show()
```

### Histograma 2D de contorno

```python
df = px.data.tips()

fig = px.density_contour(df, x="total_bill", y="tip")
fig.update_traces(contours_coloring="fill", contours_showlabels = True)
fig.show()
```
