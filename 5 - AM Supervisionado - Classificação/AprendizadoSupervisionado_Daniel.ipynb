##KNN

Já com o Método de Classificação K-NN, é executado um calculo para medir a distância entre os dados para fazer sua classificacão, dessa forma é possıvel classificar o quão próximo é um dado do outro. Dessa forma, esse método não necessita de treinamento e é incremental. Por outro lado, seu armazenamento total de todo os dados de treino e sua consulta são proporcionais ao custo computacional.

###Passo 1: Treinamento

Basicamente, ele armazena todos os exemplos do conjunto de treinamento, juntamente com suas classes (no caso de classificação)

###Passo 2: Medição de distância

O k-NN usa uma medida de distância, como a distância euclidiana, para calcular a proximidade entre as instâncias de treinamento e a nova instância que precisa ser classificada ou prevista.

###Passo 3: Escolha do parâmetro k

O parâmetro "k" no k-NN especifica o número de vizinhos mais próximos que serão considerados ao tomar uma decisão. A escolha do valor de "k" é crucial, pois afeta a qualidade das previsões. Um valor muito pequeno pode ser sensível a ruído nos dados, enquanto um valor muito grande pode suavizar demais as decisões.

###Passo 4: Classificação

No caso de classificação, para prever a classe de uma nova instância, o algoritmo encontra os "k" exemplos mais próximos a ela com base na distância e verifica qual classe é mais frequente entre esses "k" vizinhos. A nova instância é atribuída à classe mais comum entre os vizinhos.

##Árvore de decisão

image.png

Método de Classificação Árvore de Decisão, elas são muito flexíveis e fáceis de entender.
Principalmente quando o problema é classificar valores categóricos ou predizer valores contínuos pois seu funcionamento é dado pelo caminho formado por pontos de decisão (nóss da árvore), criando-se ramos para sua classificacão. Um fator notório da árvore é que o algoritmo cria os próprios classificadores sem nenhum ajuste anterior.

Exemplo

Pode-se utilizar para critério de uma árvore de decisão diversos tipos de índices, como Gini, Entropia, Variância.

Para evitar overfitting, seria interessante também realizar uma validação cruzada

```
Imagine que temos um exemplo se a pessoa irá ou não ao mercado

Condição do Tempo:
- Ensolarado
  - Temperatura:
    - Quente
      - Vai
    - Moderada
      - Vai
    - Fria
      - Não vai
- Nublado
  - Não vai
- Chuvoso
  - Temperatura:
    - Quente
      - Vai
    - Moderada
      - Não vai
    - Fria
      - Não vai
```


image.png

Basicamente o passo a passo a ser realizado é:

Passo 1: Calcular o índice de Gini para o nó "Ensolarado" (antes da divisão):
```
Gini(Ensolarado) = Jogar(Ensolarado)/Total(Ensolarado)
Gini(Ensolarado) = NãoJogar(Ensolarado)/Total(Ensolarado)
```
Passo 2: Calcular o índice de Gini para o nó "Quente" (Depois de escolher ensolarado):

```
Gini(Ensolarado|Quente) = Jogar(Ensolarado|quente)/Total(Ensolarado|quente)
Gini(Ensolarado|Quente) = NãoJogar(Ensolarado|quente)/Total(Ensolarado|quente)
```
Passo 3: Calcular o índice de Gini para o nó "Moderado" (Depois de escolher ensolarado):
```
Gini(Ensolarado|Quente) = Jogar(Ensolarado|Moderado)/Total(Ensolarado|Moderado)
Gini(Ensolarado|Quente) = NãoJogar(Ensolarado|Moderado)/Total(Ensolarado|Moderado)
```

Passo 4: Calcular o índice de Gini para o nó "Frio" (Depois de escolher ensolarado):
```
Gini(Ensolarado|Quente) = Jogar(Ensolarado|Frio)/Total(Ensolarado|Frio)
Gini(Ensolarado|Quente) = NãoJogar(Ensolarado|Frio)/Total(Ensolarado|Frio)
```

Passo 5...: Calcular para os outros nós Nublado e Chuvoso também

Nesta árvore de decisão, você escolheria o nó com o menor índice de Gini como ponto de divisão. Em cada folha (nó terminal), a classe predominante é a que a árvore de decisão irá prever para uma nova instância que siga os ramos correspondentes.

##RandomForest

É baseada na árvore de decisão

Árvores de decisão são:

*   Fáceis de fazer
*   Fáceis de entender
*   Fáceis de implementar

Entretanto, na prática não são também:

*   Boas para se analisar





Random forests por sua praticidade e flexibilidade para aumentar sua acurácia.


###Passo 1: Bootstrap

Para iniciar é necessário criar um bootstrap dataset, nada mais do que uma amostragem de dados:

Para construir cada árvore na floresta, o Random Forest faz uma amostragem com reposição (bootstrap) dos dados de treinamento. Isso significa que ele cria um conjunto de treinamento para cada árvore, selecionando aleatoriamente exemplos dos dados originais. Isso resulta em diferentes conjuntos de treinamento para cada árvore, o que introduz diversidade e ajuda a reduzir o overfitting.

###Passo 2: Construir a árvore de decisão

Para cada um dos conjuntos de treinamento amostrados, uma árvore de decisão é construída. No entanto, durante a construção de cada árvore, apenas um subconjunto das características (atributos), ou seja, apenas algumas colunas escolhidas, **de forma aleatória, são consideradas para divisão de cada nó.** Isso introduz mais aleatoriedade e reduz a tendência das árvores de se tornarem altamente correlacionadas.

Para saber qual é o subconjunto mais otimizado para tal seleção, é necessário:


*   Passo 1: Escolher um subconjunto
*   Passo 2: Testar a RandomForest
*   Passo 3: Voltar ao Passo 1, mudando a qtd de atributos selecionados

Geralmente se utiliza a raíz da quantidade de variáveis para iniciar a randomforest




Dessa forma, faz-se diversas árvores de decisão para poder analisa-las mais tarde. Criando-se assim uma floresta aleatoria.

###Passo 3: Votação ou média das previsões

Depois de todas as árvores serem construídas, quando uma nova instância precisa ser classificada (ou prevista no caso de regressão), cada árvore faz sua própria previsão.

Para problemas de classificação, as previsões de cada árvore são combinadas por meio de um processo de votação. A classe mais comum entre as previsões individuais das árvores é escolhida como a **previsão final da floresta.**

Para problemas de regressão, as previsões individuais das árvores são combinadas tirando a média, resultando em uma previsão final.

###Redução de Overfitting


A combinação das previsões de várias árvores reduz a tendência de overfitting, pois a média ou a maioria das previsões torna-se mais estável e menos sensível a variações nos dados de treinamento.

###Como saber se a randomforest é boa

Geralmente muitas tuplas são ignoradas quando as árvores de decisão são geradas. Dessa forma, elas vão indo para uma espécie 'fora da sacola'. No final, quando a randomforest foi gerada, ela é testada nessas tuplas que não foram utilizadas. Assim, dá para se ter um parâmetro do quão acurado a randomforest foi em proporção aos 'fora da sacola'.

##XGSBOOST

O XGBoost é um algoritmo altamente eficaz para problemas de classificação, pois aproveita as vantagens das árvores de decisão enquanto gerencia o overfitting por meio de técnicas de regularização e gradient boosting. Sua combinação de várias árvores fracas resulta em um modelo forte e robusto.

###Passo 1:

O processo do XGBoost começa com um modelo inicial simples. Geralmente, esse modelo inicial é apenas uma estimativa global, como a média das classes no conjunto de treinamento.

###Passo 2: Cálculo dos Resíduos e Gradientes (LOSS FUNCTION)

Para cada instância no conjunto de treinamento, são calculados os resíduos, que representam o erro entre a previsão atual do modelo e o valor verdadeiro da classe. Além disso, são calculados os gradientes da função objetivo em relação a esses resíduos. O gradiente indica a direção na qual a função objetivo está diminuindo mais rapidamente.

image.png

###Passo 3: Construção da árvore de decisão

O XGBoost começa a construir uma árvore de decisão para lidar com os resíduos e gradientes calculados. Ele otimiza as divisões nos nós da árvore usando os gradientes e hessianos para guiar as decisões. Cada nó é dividido na direção que reduz o erro da função objetivo. A árvore é construída de maneira incremental, adicionando nós um a um.

image.png

###Passo 4: Regularização e Função Objetivo

O XGBoost utiliza termos de regularização na função objetivo para evitar overfitting. Esses termos penalizam modelos mais complexos, incentivando a construção de árvores mais rasas. A função objetivo é a soma ponderada dos erros, dos termos de regularização e de outros componentes que capturam a complexidade da árvore.

###Passo 5: Atualização do modelo

Após a construção da primeira árvore, o modelo é atualizado. A previsão do modelo inicial é ajustada pelos resíduos da primeira árvore. Isso corrige parte do erro do modelo inicial.

###Passo 6: Iteração das árvores

O processo de construção e atualização é repetido para várias iterações. Cada nova árvore tenta corrigir os erros cometidos pelas árvores anteriores. O modelo final é uma combinação ponderada das previsões de todas as árvores individuais.

###Passo 7: Previsão Final

Para fazer previsões para novas instâncias, o XGBoost percorre todas as árvores construídas e somas as previsões de cada uma delas. A classe prevista é geralmente determinada pela classe majoritária das previsões.

##Sensibilidade ou Recall

A sensibilidade mede a proporção de instâncias positivas que foram corretamente identificadas pelo modelo em relação ao número total de instâncias positivas verdadeiras. Essa métrica é importante quando o foco está em minimizar os falsos negativos. Também é chamada de taxa de verdadeiros positivos ou recall.

##Acurácia

A acurácia mede a proporção de instâncias corretamente classificadas em relação ao número total de instâncias. É uma métrica simples e intuitiva, mas pode ser enganadora quando as classes estão desequilibradas. Por exemplo, se a maioria das instâncias pertence a uma classe, um modelo que sempre prevê essa classe pode ter alta acurácia, mas não é necessariamente bom em identificar a classe minoritária.

##Precisão

A precisão mede a proporção de instâncias positivas (classe prevista como positiva) que são realmente positivas (classe verdadeira). Em outras palavras, é uma métrica que avalia o quanto das previsões positivas do modelo é verdadeiro positivo. A precisão é útil quando o foco está em minimizar os falsos positivos.

##F1 Score

O F1 Score é uma métrica que combina precisão e sensibilidade em uma única medida. Ele é a média harmônica dessas duas métricas e é útil quando você deseja equilibrar a importância tanto dos verdadeiros positivos quanto dos falsos positivos/negativos. O F1 Score tende a ser mais adequado quando as classes estão desequilibradas.

###Resumo

*   Acurácia: Mede a proporção de instâncias corretamente classificadas em relação ao total. Pode ser enganosa com classes desequilibradas.
*   Precisão: Mede a proporção de instâncias classificadas como positivas que são realmente positivas.
*   Sensibilidade: Mede a proporção de instâncias positivas corretamente identificadas.
*   F1 Score: Combina precisão e sensibilidade em uma métrica única que é útil quando se deseja considerar tanto os verdadeiros positivos quanto os falsos positivos/negativos de maneira equilibrada.
